### 1. Load our PDF Document

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader(PDF_PATH)

loader
```

```python
data = loader.load()

len(data)
```

### 2. Split data into **chunks**

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    separators=['\n', '.'],
    chunk_size=500,
    chunk_overlap=100,
    length_function=len
)

chunks = splitter.split_documents(data)

len(chunks)
```

```python
print(chunks[0])
```

### 3. Create Chunks and Vector Store
1. Convert each chunk into embedding.
2. Store these embeddings into vector store.

```python
from re import search
from langchain_chroma import Chroma
from langchain_community.embeddings import HuggingFaceEmbeddings


embedding_model = HuggingFaceEmbeddings(model_name='intfloat/multilingual-e5-small')

# Check if Chroma Vector DB exists.
if os.path.exists(CHROMA_DIR):
    print("Load DB from Drive....")
    vector_store = Chroma(
        persist_directory=CHROMA_DIR,
        embedding_function=embedding_model
    )
else:
    print("Build DB from scratch....")
    vector_store = Chroma.from_documents(
        documents=chunks,
        embedding=embedding_model,
        persist_directory=CHROMA_DIR
    )



retriever = vector_store.as_retriever(
    search_type='mmr',
    search_kwargs={"k":6, "fetch_k": 10}
)
```

### 4. System Prompt & RAG Chain

```python
# Instantiate our Open-Source Model
from langchain_groq import ChatGroq

model = ChatGroq(
        model="openai/gpt-oss-120b",
        temperature = 0.2
)
```

```python
from langchain_core.prompts import ChatPromptTemplate

system_prompt = """
You are a customer service representative for an electronics company.

Use only the information provided in the text below.

If the information is missing, say you don't know and suggest contacting human support.

Match the user's language (Arabic or English).

Keep your responses short and clear.

{context}
""".strip()

prompt = ChatPromptTemplate.from_messages([
    ("system", system_prompt),
    ("human", "{query}")
])
```

```python
from langchain_core.runnables import RunnablePassthrough

def format_docs(docs):
  """
  Format retrieved documents
  """
  return "\n\n".join(d.page_content for d in docs)


rag_chain = (
    {'context': retriever | format_docs, "query":RunnablePassthrough()}
    | prompt
    | model
)
```

### 5. Run the RAG via Gradio UI

```python
import gradio as gr

def respond(message, chat_history):
  """
  This function is called when 'submit' is pressed.
  """

  # 1. Check that the message is not empty.
  if not message or not message.strip():
    return chat_history, ""

  # 2. Invoke RAG Chain with only the current question.
  try:
    response = rag_chain.invoke(message)
    bot_reply = response.content
  except Exception as e:
    bot_reply = f"Sorry, An Error Occurred: {e}."

  # 3. Add the new question and answer to the chat history in 'messages' format.
  if chat_history is None:
    chat_history = []

  # Add user's message
  chat_history.append({"role": "user", "content":message})
  # Add Bot Reply
  chat_history.append({"role": "assistant", "content":bot_reply})

  return chat_history, ""


with gr.Blocks() as demo:
  gr.Markdown("## Customer Support Chatbot")

  chatbot = gr.Chatbot(label="Customer Support AI Assistant.", type="messages")

  msg = gr.Textbox(
      placeholder= "Type your question (arabic/english)...",
      label="Your Question")
  clear = gr.Button("Clear Chat")

  # Link the 'respond' function to the interface.
  msg.submit(respond, [msg, chatbot], [chatbot, msg])

  # Link the 'clear' button.
  clear.click(lambda: ([], ""), None, [chatbot, msg])


demo.launch(share=True)
```
